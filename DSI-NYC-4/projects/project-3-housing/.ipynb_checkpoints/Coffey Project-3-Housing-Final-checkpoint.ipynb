# INTRODUCTION

This notebook explores home price in Ames, Iowa in an effort to build a predictive linear regression model. 

The model will explore a variety of numerical and categorical variables that describe homes sold in Ames from 2006 - 2010, and will settle on the model with the highest R2 after running the model through Lasso.

In addition to modeling home sales in Ames, this investigation performs extensive data analysis on  home sales by neighborhood, square footage, and quality.  It also includes hypothesis testing of mean home sale prices by quality and square footage.

It is the strong suspicion of this author that neighborhood is not by itself the foremost predictor of home values.

## Hypothesis and ANOVA Tests: 

This testing attempts to reveal three insights: 

--Is there a meaningful difference between the mean sale prices
of homes at different quality levels? 


--Is there a statistically significant difference in sales prices as quality increases?


--Is there a relationship between price and above ground square footage?

```python
#Creating new variables that break-out the quality levels. 

cat1_0 = df[df['Quality']== 1]
cat1_1 = df[df['Quality']== 2]
cat1 = df[df['Quality']== 3]
cat2 = df[df['Quality']==4]
cat3 = df[df['Quality']==5]
cat4 = df[df['Quality']==6]
cat5 = df[df['Quality']==7]
cat6 = df[df['Quality']==8]
cat7 = df[df['Quality']==9]
cat8 = df[df['Quality']==10]
```

ANOVA TEST: Sales Price relative to the Quality of a home.

```python
cw_lm=ols('SalePrice ~ Quality', data=df).fit() #Specify C for Categorical
print(sm.stats.anova_lm(cw_lm, typ=2))
```

There is a difference in the mean sales price when comparing homes at
different quality levels. 

Two Sided Hypothesis Test: Are there statistically significant differences in mean price when comparing 
homes at adjoining quality levels (i.e. Quality 5 v Quality 6, Quality 6 v Quality 7, etc.)

```python

from scipy.stats import ttest_ind
tstat, pvalue = ttest_ind(cat1['SalePrice'], cat2['SalePrice'])
print('t-stat Quality Level 3 & 4: %s' %tstat)
print'p-value: %s' %pvalue
```

H0: The mean sales price of homes at quality levels 3 & 4 are the same.

HA: The mean sales price of homes at adjoining quality levels 3 & 4 are not same (Two Sided).

Reject the null hypothesis

```python
#Reject null that mean in Quality level 4 = Quality Level 5

tstat, pvalue = ttest_ind(cat2['SalePrice'], cat3['SalePrice'])
print('t-stat Quality Level 4 & 5: %s' %tstat)
print('p-value %s' %pvalue)
```

H0: The mean sales price of homes at quality levels 4 & 5 are the same.

HA: The mean sales price of homes at adjioning quality levels 4 & 5 are not same (Two Sided).

Reject the null hypothesis

```python
#Reject null that mean in Quality level 5 = Quality Level 6
tstat, pvalue = ttest_ind(cat3['SalePrice'], cat4['SalePrice'])
print('t-stat Quality Level 5 & 6: %s' %tstat) 
print ('p-value %s' %pvalue)
```

## Hypothesis testing the relationship between quality and square footage.

Testing whether the the mean square footage of a home is different at the various quality levels

```python
tstat, pvalue = ttest_ind(cat1_1['AbvGroundLiv'], cat2['AbvGroundLiv'])
print('Q. Level 2 & 3 (Reject Null):', tstat , 'p-value', pvalue)

tstat, pvalue = ttest_ind(cat1['AbvGroundLiv'], cat2['AbvGroundLiv'])
print('Q. Level 3 & 4(Fail to Reject):', tstat , 'p-value', pvalue)

tstat, pvalue = ttest_ind(cat2['AbvGroundLiv'], cat3['AbvGroundLiv'])
print('Q. Level 4 & 5 (Reject Null): ', tstat , 'p-value', pvalue)

tstat, pvalue = ttest_ind(cat3['AbvGroundLiv'], cat4['AbvGroundLiv'])
print('Q. Level 5 & 6:(Reject Null)', tstat , 'p-value', pvalue)

tstat, pvalue = ttest_ind(cat4['AbvGroundLiv'], cat5['AbvGroundLiv'])
print('Q. Level 6 & 7:(Reject Null)', tstat , 'p-value', pvalue)

tstat, pvalue = ttest_ind(cat5['AbvGroundLiv'], cat6['AbvGroundLiv'])
print('Q. Level 7 & 8:(Reject Null)', tstat , 'p-value', pvalue)

tstat, pvalue = ttest_ind(cat6['AbvGroundLiv'], cat7['AbvGroundLiv'])
print('Q. Level 8 & 9:(Fail to Reject)', tstat , 'p-value', pvalue)

tstat, pvalue = ttest_ind(cat7['AbvGroundLiv'], cat8['AbvGroundLiv'])
print('Q. Level 9 & 10:(Reject Null)', tstat , 'p-value', pvalue)
```

As we see from the hypothesis tests above, there is not a statistically significant difference in square footage at quality levels 8 & 9, and quality levels 3 & 4. When buying a home at quality level 8, one might be getting the same square footage as a home at quality level 9. 

# LINEAR REGRESSION

We know that there are a few categories that correlate to price, including quality and above ground living square footage. Location may have an influence too. Let's build a full model and attempt to run some predictions.

```python
list(df.columns)
```

Building new dummie variables based on location, building type, style, roodstyle and month sold. 

```python
df = df.drop('Id', axis = 1)

df = pd.concat([
        df.drop([
                'Hood', 
                'BldgType', 
                'Style', 
                'RoofStyle', 
                'MonthSold', 
                'month_sold2']
                , axis = 1), 
    pd.get_dummies(df[['Hood', 'BldgType', 'Style', 'RoofStyle', 'MonthSold']],\
    drop_first=True)], axis = 1)
```

```python
df.columns
```

```python
# Full Model with One Categorical Neighborhood Variable
import statsmodels.api as sm
from statsmodels.formula.api import ols
from sklearn import linear_model
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
y = df['SalePrice']

X = df.drop('SalePrice', axis = 1)

X_scaled = pd.DataFrame(MinMaxScaler().fit_transform(X))

X_scaled.columns = X.columns

X_scaled.head()
```

```python
sns.heatmap(X_scaled.corr())
```

The above heatmap shoes that correlations between many of our dependent variables. 

## FULL MODEL

Running a full linear model with a test/train split

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.33,  random_state = 22)

lm = LinearRegression(fit_intercept = True, normalize = True)

lm.fit(X_train, y_train)
```

```python
y_predict = lm.predict(X_test)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
from sklearn.metrics import r2_score
y_trueTest = y_test
y_predTest = lm.predict(X_test)

lr_r2Split =  r2_score(y_true=y_trueTest, y_pred=y_predTest)
print "R2 Test Set:", lr_r2Split
```

### Conclusion: The full model has an R2 of .86. It has decent predictive ability 

Let's examine the coefficients to see whether some are stronger than others

We can build a decent linear regression model predicting home prices with  just five variables. Nonetheless, the model with the highest R2 includes all variables, with dummified neighborhood variables. 

It's worth exploring whether increasing the complexity of the model through the dummification of every variable will increase the R2. My suspicion is that it does. 

While location is an important part of this model, as suspected, it is less important as above ground square footage. Other variables such as the quality of a home and the year built also have strong predictive abilities.

Still, the best model must include location. That being said, the locations with the highest influence on home price are limited to whether a home is in two specific high priced neighborhoods.  

If the goal going forward is to build a model that specifically predicts sales in Ames, it makes sense to include every neighborhood in the model. If the goal is to build a model with general predictive abilities, it may make more sense to include quality, year built and square footage, and then fine tune the model for specific cities.

At the end of the day, these findings are not experimental, and are generalizable to the population. Given that there is no random sampling beyond Ames, the model for now is generalizable to the Ames area.
